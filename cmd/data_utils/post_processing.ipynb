```python
# Write the output file 

with open("classes-train-classif.txt", "w") as filout:
    name_file_count = set()
    # go through the panda_frame:
    for name_file in df3.source: 
        # new label set for each file
        label_file = set()
        if name_file not in name_file_count:
            name_file_count.add(name_file)
            for label in df3[df3.source == name_file].level_1:
                label = label[6:]
                print(name_file, label)
                that_line = [name_file, label, 'term']
                that_line_txt = "\t".join(that_line)
                filout.write("{}\n".format(that_line_txt))
```
